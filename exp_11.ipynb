```r
import matplotlib.pyplot as plt
import numpy as np# importing packages
import seaborn as sns

# loading dataset
data = sns.load_dataset("iris")

# draw lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

np.random.seed(0)
ratings = np.random.randint(1, 11, size=100)
plt.figure(figsize=(8, 6))
plt.hist(ratings, bins=range(1, 12), edgecolor='black', alpha=0.7)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Movie Ratings')
plt.xticks(range(1, 11))
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```r

library(ggplot2)
set.seed(123)
data <- data.frame(
  Height = rnorm(100, mean = 170, sd = 10),
  Weight = rnorm(100, mean = 70, sd = 10)
)
ggplot(data, aes(x = Height, y = Weight)) +
  geom_point() +
  geom_smooth(method = "lm", se = FALSE) +
  labs(
    title = "Relationship Between Height and Weight",
    x = "Height (cm)",
    y = "Weight (kg)"
  ) +
  theme_minimal()
```
